* __SkyEng__ - онлайн-школа по изучению английского языка, предлагает индивидуальные уроки с носителями языка, доступ к разнообразным обучающим материалам и гибкость в выборе расписания. Обучение происходит на собственной учебной платформе Vimbox, а также при помощи мобильного приложения, расширений для браузера и обучающих рассылок.
* __Lingvoleo__ - это образовательная платформа для изучения и практики иностранного языка, построенная на игровой механике. Онлайн-тренировки по грамматике и лексике позволяют изучать английский с нуля. Она включает различные уровни сложности, интерактивные упражнения, игры и функцию повторения.
   
Основываясь на анализе лендингов SkyEng и Lingualeo, а также учитывая общепринятые лучшие практики, можно предложить следующие гипотезы для улучшения первого экрана лендинга:

1. __Установить яркий заголовок:__ Заголовок должен четко и кратко передавать основное преимущество продукта. Подзаголовок может дать дополнительную информацию или уточнить предложение. Призыв к действию должен быть ярким и заметным.
2. __Внедрить анимацию:__ Анимации могут привлечь внимание и улучшить пользовательский опыт, но их следует использовать с умом, чтобы не отвлекать внимание.
3. __Включить формы для сбора контактных данных:__ Форма для сбора контактных данных должна быть видна выше складки, чтобы упростить процесс конверсии.
5. __Улучшить скорость загрузки:__ Быстрая загрузка страницы важна для удержания внимания посетителей.
6. __Вклюить элементы доверия:__ Отзывы клиентов, сертификаты, награды и другие элементы доверия могут увеличить доверие посетителей.
7. __Улучшить качество изображенй:__ Качественные изображения и видео могут улучшить впечатление от сайта.

*Пример гипотезы:*

#### Кейс №4
Вы решили сравнивать метрику CPA в двух группах. Размер выборки - 2350 элементов в каждой группе.<p> 
Для проверки нормальности распределения на выборке в 2350 наблюдений применили,критерий Шапиро-Уилка и получили p-value, равный 0.00002, alpha = 5%.
- Какой бы вывод мы могли сделать в данном случае?
- В этом случае какой статистический критерий для проверки первоначальной гипотезы тут лучше всего подойдёт и почему ?

__Вывод:__

Исходя из полученного р-value = 0.00002, который меньше уровня значимости alpha = 5%, можно сделать вывод о том, что распределение данных не является нормальным.
В данном случае, учитывая ненормальность распределения, для проверки первоначальной гипотезы лучше всего подойдет непараметрический критерий, например, критерий Манна-Уитни. Этот критерий не требует предложения о нормальности данных и может быть использован для сравнения двух независимых выборок.
